```python
def selection_sort(arr):
    n = len(arr)
    for i in range (n-1):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index=j
        temp = arr[i]        
        arr[i] = arr[min_index]
        arr[min_index] = temp
    print("Array After Selection Sort")
    print(*arr)
n = int(input("Enter the size of the array"))    
arr = []
print("Enter the elements")
for i in range(n):
    element = int(input())
    arr.append(element)
    
print("Array Before Selection Sort")
print(*arr)
selection_sort(arr)
```
